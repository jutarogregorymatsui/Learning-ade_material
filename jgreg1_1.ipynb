## １．多項式（１次関数）
### $$ y = 4x+3$$

```python
x = np.linspace(-5,5,11)
y = 4*x + 3
plt.plot(x,y)

plt.grid()
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.xlabel('$X$', fontsize = 16)
plt.ylabel('$Y$', fontsize = 16)
```

## １．多項式（２次~５次関数）

```python
fig = plt.figure(figsize=(11,7))

ax2 = fig.add_subplot(221)
ax3 = fig.add_subplot(222)
ax4 = fig.add_subplot(223)
ax5 = fig.add_subplot(224)

x2 = np.linspace(-7, 5, 100)
y2 = 3*x2**2 + 5*x2 - 10
ax2.plot(x2, y2)

ax2.grid()
ax2.axhline(0, color='gray')
ax2.axvline(0, color='gray')
ax2.set_ylabel('$Y$', fontsize = 14)
ax2.set_title('$y= 3x^2 + 5x -10$')


x3 = np.linspace(-5, 5, 100)
y3 = 3*x3**3 - 5*x3**2 - 30*x3 + 200
ax3.plot(x3, y3)

ax3.grid()
ax3.axhline(0, color='gray')
ax3.axvline(0, color='gray')
ax3.set_title('$y= 3x^3 - 5x^2 -30x + 200 $')


x4 = np.linspace(-5, 11, 100)
y4 = 3*x4**4 - 30*x4**3 - 30*x4**2 + 300*x4 + 4
ax4.plot(x4, y4)

ax4.grid()
ax4.axhline(0, color='gray')
ax4.axvline(0, color='gray')
ax4.set_xlabel('$X$', fontsize = 14)
ax4.set_ylabel('$Y$', fontsize = 14)
ax4.set_title('$y= 3x^4 - 30x^3 - 30x^2 + 300x + 4$')


x5 = np.linspace(-5, 9, 100)
y5 = 3*x5**5 - 30*x5**4 - 40*x5**3 + 650*x5**2 + 40*x5 + 100
ax5.plot(x5, y5)

ax5.grid()
ax5.axhline(0, color='gray')
ax5.axvline(0, color='gray')
ax5.set_xlabel('$X$', fontsize = 14)
ax5.set_title('$y= 3x^5 - 30x^4 -40x^3 + 650x^2 + 40x + 100$')

plt.show()
```

## ２．三角関数

```python
fig = plt.figure(figsize=(12,3))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y1 = np.sin(x)
ax1.plot(x, y1)

ax1.grid()
ax1.set_title('$y=sin(x)$')
ax1.axhline(0, color='gray')
ax1.axvline(0, color='gray')


y2 = np.cos(x)
ax2.plot(x, y2)

ax2.grid()
ax2.set_title('$y=cos(x)$')
ax2.axhline(0, color='gray')
ax2.axvline(0, color='gray')


y3 = np.tan(x)
ax3.scatter(x, y3, s=2)

ax3.grid()
ax3.set_ylim(-10,10)
ax3.set_title('$y=tan(x)$')
ax3.axvline(np.pi/2, color='gray')
ax3.axvline(-np.pi/2, color='gray')
ax3.axvline(3*np.pi/2, color='gray')
ax3.axvline(-3*np.pi/2, color='gray')
ax3.axhline(0, color='gray')
ax3.axvline(0, color='gray')

plt.show()
```

## ３．分数関数

```python
fig = plt.figure(figsize=(11,7))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

x1 = np.linspace(-5, 5, 1000)
y1 = x1/(x1 -1)
ax1.scatter(x1, y1, s=1)

ax1.set_ylim(-10,10)
ax1.grid()
ax1.axhline(0, color='gray')
ax1.axvline(0, color='gray')
ax1.axvline(1, color='gray')
ax1.axhline(1, color='gray')
ax1.set_ylabel('$Y$', fontsize = 12)
ax1.set_title('$y= x/(x-1)$')


x2 = np.linspace(-5, 5, 1000)
y2 = 8/(4 - x2**2)
ax2.scatter(x2, y2, s=2)

ax2.set_ylim(-10,10)
ax2.grid()
ax2.axhline(0, color='gray')
ax2.axvline(0, color='gray')
ax2.axvline(-2, color='gray')
ax2.axvline(2, color='gray')
ax2.set_ylabel('$Y$', fontsize = 12)
ax2.set_title('$y= 8/(4-x^2)$')


x3 = np.linspace(-5, 8, 1000)
y3 = (x3**2 - 3)/(2*x3 - 4)
ya = 1/2*x3 + 1

ax3.scatter(x3, y3, s=2)
ax3.plot(x3, ya, color='gray')
ax3.set_ylim(-3,6)
ax3.grid()
ax3.axhline(0, color='gray')
ax3.axvline(0, color='gray')
ax3.set_xlabel('$X$', fontsize = 12)
ax3.set_ylabel('$Y$', fontsize = 12)
ax3.set_title('$(x^2 - 3)/(2x - 4)$')


x4 = np.linspace(-5, 5, 1000)
y4 = 2/x4 + 6*x4**2
ax4.scatter(x4, y4, s=2)

ax4.set_ylim(-100,100)
ax4.grid()
ax4.axhline(0, color='gray')
ax4.axvline(0, color='gray')
ax4.set_xlabel('$X$', fontsize = 12)
ax4.set_ylabel('$Y$', fontsize = 12)
ax4.set_title('$y= 2/x + 6x^2$')
```

## ４．双曲線関数

```python
fig = plt.figure(figsize=(12,3))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y1 = np.sinh(x)
ax1.plot(x, y1)

ax1.grid()
ax1.set_ylim(-10, 10)
ax1.axhline(0, color='gray')
ax1.axvline(0, color='gray')
ax1.set_title('$y=sinh(x)$')


y2 = np.cosh(x)
ax2.plot(x, y2)

ax2.grid()
ax2.set_ylim(-10, 10)
ax2.axhline(0, color='gray')
ax2.axvline(0, color='gray')
ax2.set_title('$y=cosh(x)$')


y3 = np.tanh(x)
ax3.plot(x, y3)

ax3.grid()
ax3.set_ylim(-2,2)
ax3.axhline(0, color='gray')
ax3.axvline(0, color='gray')
ax3.set_title('$y=tanh(x)$')

plt.show()
```

## ５．指数関数、対数関数
### $$y=e^x\hspace{20mm}y=Ln(x)$$ 

```python
fig = plt.figure(figsize=(9,3))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

x1 = np.linspace(-3, 3, 1000)
y1 = np.exp(x1)
ax1.plot(x1, y1)

ax1.grid()
ax1.axhline(0, color='gray')
ax1.axvline(0, color='gray')
ax1.set_title('$y=e^x$')


x2 = np.linspace(0.001, 3, 100)
y2 = np.log(x2)
ax2.plot(x2, y2)

ax2.grid()
ax2.set_ylim(-3, 3)
ax2.axhline(0, color='gray')
ax2.axvline(0, color='gray')
ax2.set_title('$y=Ln(x)$')

plt.show()
```
